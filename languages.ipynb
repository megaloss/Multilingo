```python
from gtts import gTTS
from io import BytesIO
from IPython.display import Audio
from tempfile import TemporaryFile
from google_trans_new import google_translator
translator = google_translator()  
from time import sleep
from IPython.core.display import display
import random
```

```python
def speak(my_text,lang):
    print (my_text)
    if isinstance(my_text, list): my_text=my_text[0]

    with BytesIO() as f:
        try:
            gTTS(text=my_text, lang=lang).write_to_fp(f)
        except:
            pass
        f.seek(0)
        return display(Audio(f.read(), autoplay=True))
```

```python
langs={ 
        'de': 'German',
        'es': 'Spanish',
        'fr': 'French',
        'cs': 'Czech',
        'pl': 'Polish',
        'uk': 'Ukrainian',
        'nl': 'Dutch',
        'ru': 'Russian',
        'en': 'English'}
langs_ord=['de','es','fr','cs','pl','uk','nl','ru','en']
```

```python
with open('eng_phrases.txt','r') as f:
    phrases=f.read().split('.')
```

```python
for i in range (10):
    phrase=random.choice(phrases)
    for lang in langs_ord:
        text=translator.translate(phrase,lang_tgt=lang, lang_src='en')
        speak(langs[lang], 'en')
        sleep(2)
        speak(text, lang)
        sleep(5)
    sleep(3)
```
